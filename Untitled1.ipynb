```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import time
from datetime import timedelta
#ÂÄíÂÖ•mnistÊï∞ÊçÆÈõÜÔºåÊ≥®ÊÑèmnistÊòØ‰∏Ä‰∏™Â§ßÊùÇÁÉ©ÔºåÈúÄË¶ÅÂØºÂÖ•Êï∞ÊçÆÁâπÂæÅÂíåÊ†áÁ≠æ
X_mnist, y_mnist = fetch_openml('mnist_784', return_X_y=True, as_frame=False,
                                parser='auto')

X_train,X_valid,X_test = X_mnist[:50000],X_mnist[50000:60000],X_mnist[60000:70000]
y_train,y_valid,y_test = y_mnist[:50000],y_mnist[50000:60000],y_mnist[60000:70000]
#ÊàëËøôÈáåÊï∞ÊçÆËøõË°å‰∫Ü‰∏ÄÂÆöÁöÑÂ§ÑÁêÜÔºåÂä†Âø´ËÆ°ÁÆóÈÄüÂ∫¶
X_train_scaled = X_train / 255.0
X_valid_scaled = X_valid / 255.0
X_test_scaled = X_test / 255.0
```

```python
# ËÆ∞ÂΩïÊï¥‰∏™ÊµÅÁ®ãÂºÄÂßãÁöÑÊó∂ÂàªÔºåËøôÊòØ‰∏Ä‰∏™ËÆ°Êó∂ÁöÑË£ÖÁΩÆ
overall_start = time.perf_counter()



rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
etc = ExtraTreesClassifier(n_estimators=100, random_state=42, n_jobs=-1)
svc = SVC(probability=True,random_state=42)
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
models = [
    ('RFM',rfc),
    ('EFC',etc),
    ('SVM',svc),
    ('KNN',knn)
]
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(
    models,
        voting = 'soft',
    weights = [1,1,2,2]
)
voting_clf.fit(X_train_scaled, y_train)
y_pre = voting_clf.predict(X_valid_scaled)
print(accuracy_score(y_valid,y_pre))


# --- ËÆ°Êó∂ÁªüËÆ°ÁªìÊùü ---
overall_end = time.perf_counter()
total_duration = overall_end - overall_start

# Â∞ÜÁßíÊï∞ËΩ¨Êç¢‰∏∫ Â∞èÊó∂:ÂàÜÈíü:Áßí ÁöÑÊ†ºÂºè
formatted_time = str(timedelta(seconds=int(total_duration)))

print("\n" + "="*30)
print(f"ÁªüËÆ°Êä•ÂëäÔºö")
print(f"ËÆ≠ÁªÉÊ†∑Êú¨Êï∞Ôºö{len(X_train_scaled)}")
print(f"ÊÄªËøêË°åËÄóÊó∂Ôºö{formatted_time}")
print("="*30)
```

```python
# 1. ‰ªéÊäïÁ•®ÈõÜÊàê‰∏≠ÊèêÂèñ„ÄêÂ∑≤ÁªèËÆ≠ÁªÉÂ•Ω„ÄëÁöÑÂ≠êÊ®°Âûã
rf_model = voting_clf.estimators_[0]
et_model = voting_clf.estimators_[1]
svm_model = voting_clf.estimators_[2]
knn_model = voting_clf.estimators_[3] 

# 2. ‰ΩøÁî®ÊèêÂèñÂá∫Êù•ÁöÑÊ®°ÂûãËøõË°åÈ¢ÑÊµã (Ê≥®ÊÑèÂèòÈáèÂêç)
y_rfc_pre = rf_model.predict(X_valid_scaled)  # ‰ΩøÁî® rf_model ËÄå‰∏çÊòØ rfc
y_etc_pre = et_model.predict(X_valid_scaled)
y_svc_pre = svm_model.predict(X_valid_scaled)
y_knn_pre = knn_model.predict(X_valid_scaled)

# 3. ËæìÂá∫ÂáÜÁ°ÆÁéá
y_valid = y_valid.astype(np.int64)#ÂéüÂßãÁöÑy_validÊòØÂ≠óÁ¨¶‰∏≤ÔºåËÄåÈ¢ÑÊµãÂá∫Êù•ÁöÑÁªìÊûúÊòØÊï∞Â≠óÔºåÂ∞ÜÂ≠óÁ¨¶‰∏≤ËΩ¨Âåñ‰∏∫Êï∞Â≠óÔºåÂ∞±ËÉΩËøõË°å‰∏ãËæπÁöÑÂáÜÁ°ÆÁéáËÆ°ÁÆó
print('RFC accuracy:', accuracy_score(y_valid, y_rfc_pre))
print('ETC accuracy:', accuracy_score(y_valid, y_etc_pre))
print('SVC accuracy:', accuracy_score(y_valid, y_svc_pre))
print('KNN accuracy:', accuracy_score(y_valid, y_knn_pre))
```

```python
#Ëøô‰∏Ä‰∏™codeÊòØ‰∏∫‰∫ÜÊúÄÂêéËÆ°ÁÆóÂáÜÁ°ÆÊÄßÊù•ËΩ¨ÂåñÁöÑÔºåÂ∞ÜÂÖ∂ÊèêÂà∞Á¨¨‰∏Ä‰∏™code‰πüÂèØ‰ª•
y_train = y_train.astype(np.int64)
y_valid = y_valid.astype(np.int64)
y_test = y_test.astype(np.int64)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
final_learner = LogisticRegression(random_state=42)

overall_start = time.perf_counter() #ËÆ°Êó∂

stacking_clf = StackingClassifier(
    estimators=models,
    final_estimator=final_learner,
    cv=3,  # number of cross-validation folds
    n_jobs=-1,
    stack_method='predict_proba'
)
stacking_clf.fit(X_train_scaled,y_train)
y_pre = stacking_clf.predict(X_valid_scaled)
score = accuracy_score(y_valid,y_pre)
print(score)

# --- ËÆ°Êó∂ÁªüËÆ°ÁªìÊùü ---
overall_end = time.perf_counter()
total_duration = overall_end - overall_start

# Â∞ÜÁßíÊï∞ËΩ¨Êç¢‰∏∫ Â∞èÊó∂:ÂàÜÈíü:Áßí ÁöÑÊ†ºÂºè
formatted_time = str(timedelta(seconds=int(total_duration)))

print("\n" + "="*30)
print(f"ÁªüËÆ°Êä•ÂëäÔºö")
print(f"ËÆ≠ÁªÉÊ†∑Êú¨Êï∞Ôºö{len(X_train_scaled)}")
print(f"ÊÄªËøêË°åËÄóÊó∂Ôºö{formatted_time}")
print("="*30)
```

```python
# ÊúÄÁªàÊµãËØïÈõÜËØÑ‰º∞
y_test_pred1 = voting_clf.predict(X_test_scaled)
final_score = accuracy_score(y_test.astype(int), y_test_pred1.astype(int))

print(f"üéâ votingÊ®°ÂûãÊúÄÁªàÊµãËØïÈõÜÂáÜÁ°ÆÁéá: {final_score:.4f}")
```

```python
# ÊúÄÁªàÊµãËØïÈõÜËØÑ‰º∞
y_test_pred2 = stacking_clf.predict(X_test_scaled)
final_score = accuracy_score(y_test.astype(int), y_test_pred2.astype(int))

print(f"üéâ stackingÊ®°ÂûãÊúÄÁªàÊµãËØïÈõÜÂáÜÁ°ÆÁéá: {final_score:.4f}")
```
